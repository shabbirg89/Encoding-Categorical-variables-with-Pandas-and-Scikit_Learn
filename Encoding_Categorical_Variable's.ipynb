```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("melb_data.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
obj_df1 = obj_df
```

```python
obj_df[obj_df.isnull().any(axis=1)]
```

```python
obj_df["CouncilArea"].value_counts()
```

```python
obj_df = obj_df.fillna({"CouncilArea": "Moreland"})
```

```python
obj_df.isnull().sum()
```

### 1 - Find and Replace

```python
obj_df["Method"].value_counts()
```

```python
cleanup_methods = {"Method": {"S": 1, "SP": 2, "PI": 3, "VB": 4,
                                  "SA": 5 }}
```

```python
obj_df.replace(cleanup_methods, inplace=True)
obj_df.head()
```

```python
obj_df.dtypes
```

### 2 - Label Encoding

```python
obj_df['Type'].value_counts()
```

```python
obj_df["Type"] = obj_df["Type"].astype('category')
obj_df.dtypes
```

```python
obj_df["Type"] = obj_df["Type"].cat.codes
obj_df.head(50)
```

```python
obj_df['Type'].value_counts()
```

### 3 - One Hot Encoding

```python
print(obj_df['CouncilArea'].value_counts())
print()
print(obj_df['Regionname'].value_counts())
```

```python
pd.get_dummies(obj_df, columns=["CouncilArea", "Regionname"], prefix=["Council", "Region"]).head(3)
```

### 4 - Custom Binary Encoding

```python
obj_df['Suburb'].value_counts()
```

```python
obj_df["Suburb_binary"] = np.where(obj_df["Suburb"].str.contains("Reservoir"), 1, 0)
```

```python
df2 = obj_df[["Suburb", "Suburb_binary"]]
```

```python
df2.query("Suburb == 'Reservoir'")
```

## Using Scikit-Learn

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
obj_df1["Type_code"] = lb_make.fit_transform(obj_df1["Type"])
obj_df1[["Type", "Type_code"]].head(11)
```

```python
from sklearn.preprocessing import LabelBinarizer

lb_Suburb = LabelBinarizer()
lb_results = lb_Suburb.fit_transform(obj_df1["Suburb"])
pd.DataFrame(lb_results, columns=lb_Suburb.classes_)
```

```python
import category_encoders as ce

# Get a new clean dataframe
obj_df = df.select_dtypes(include=['object']).copy()

# Specify the columns to encode then fit and transform
encoder = ce.backward_difference.BackwardDifferenceEncoder(cols=["Method"])
encoder.fit(obj_df, verbose=1)

encoder.transform(obj_df).head()
```

```python
# Get a new clean dataframe
obj_df = df.select_dtypes(include=['object']).copy()

# Specify the columns to encode then fit and transform
encoder = ce.polynomial.PolynomialEncoder(cols=["Method"])
encoder.fit(obj_df, verbose=1)

encoder.transform(obj_df).head()
```
